```python
import csv
import os

#2 Function to load current inventory from the CSV file
def load_inventory(filename='inventory.csv'):
    inventory = {}
    if os.path.exists(filename):
        with open(filename, mode='r') as file:
            csv_reader = csv.DictReader(file)
            for row in csv_reader:
                item_id = int(row['Item ID'])
                inventory[item_id] = {
                    'Item Name': row['Item Name'],
                    'Category': row['Category'],
                    'Quantity': int(row['Quantity']),
                    'Unit Price': float(row['Unit price']),
                    'Reorder Level': int(row['Reorder level']),
                }
    return inventory

# Function to save the inventory to the CSV file
def save_inventory(inventory, filename='inventory.csv'):
    with open(filename, mode='w', newline='') as file:
        fieldnames = ['Item ID', 'Item Name', 'Category', 'Quantity', 'Unit price', 'Reorder level']
        csv_writer = csv.DictWriter(file, fieldnames=fieldnames)
        csv_writer.writeheader()
        for item_id, details in inventory.items():
            row = {
                'Item ID': item_id,
                'Item Name': details['Item Name'],
                'Category': details['Category'],
                'Quantity': details['Quantity'],
                'Unit price': details['Unit Price'],
                'Reorder level': details['Reorder Level'],
            }
            csv_writer.writerow(row)

# Function to add a new item
def add_item(inventory):
    item_id = int(input("Enter Item ID: "))

    # Check for uniqueness of Item ID
    if item_id in inventory:
        print("Error: Item ID already exists.")
        return

    item_name = input("Enter Item Name: ")
    category = input("Enter Category: ")

    while True:
        try:
            quantity = int(input("Enter Quantity: "))
            if quantity < 0:
                raise ValueError("Quantity cannot be negative.")
            break
        except ValueError as e:
            print(e)

    while True:
        try:
            unit_price = float(input("Enter Unit Price: "))
            if unit_price < 0:
                raise ValueError("Unit Price cannot be negative.")
            break
        except ValueError as e:
            print(e)

    reorder_level = int(input("Enter Reorder Level: "))

    # Add to inventory and save
    inventory[item_id] = {
        'Item Name': item_name,
        'Category': category,
        'Quantity': quantity,
        'Unit Price': unit_price,
        'Reorder Level': reorder_level,
    }
    save_inventory(inventory)
    print("Item added successfully!")

# Function to update an item
def update_item(inventory):
    item_id = int(input("Enter Item ID to update: "))

    if item_id not in inventory:
        print("Error: Item ID not found.")
        return

    print(f"Current details: {inventory[item_id]}")
    while True:
        try:
            new_quantity = int(input("Enter new Quantity (or leave blank to keep current): ") or inventory[item_id]['Quantity'])
            if new_quantity < 0:
                raise ValueError("Quantity cannot be negative.")
            break
        except ValueError as e:
            print(e)

    while True:
        try:
            new_unit_price = float(input("Enter new Unit Price (or leave blank to keep current): ") or inventory[item_id]['Unit Price'])
            if new_unit_price < 0:
                raise ValueError("Unit Price cannot be negative.")
            break
        except ValueError as e:
            print(e)

    new_reorder_level = int(input("Enter new Reorder Level (or leave blank to keep current): ") or inventory[item_id]['Reorder Level'])

    # Update inventory
    inventory[item_id]['Quantity'] = new_quantity
    inventory[item_id]['Unit Price'] = new_unit_price
    inventory[item_id]['Reorder Level'] = new_reorder_level
    save_inventory(inventory)
    print("Item updated successfully!")

# Function to delete an item
def delete_item(inventory):
    item_id = int(input("Enter Item ID to delete: "))

    if item_id not in inventory:
        print("Error: Item ID not found.")
        return

    del inventory[item_id]
    save_inventory(inventory)
    print("Item deleted successfully!")


# Main Menu function
def main():
    inventory = load_inventory()

    while True:
        print("\n--- Inventory Management ---")
        print("1. Add New Item")
        print("2. Update Inventory")
        print("3. Delete Item")
        print("4. Exit")

        choice = input("Choose an option (1-4): ")

        if choice == '1':
            add_item(inventory)
        elif choice == '2':
            update_item(inventory)
        elif choice == '3':
            delete_item(inventory)
        elif choice == '4':
            print("Exiting the program.")
            break
        else:
            print("Invalid option, please choose again.")

if __name__ == "__main__":
    main()
```
